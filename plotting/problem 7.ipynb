```python
from pyarma import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "prob7_single_psum"
probsum1 = cx_mat()
probsum1.load(path)

path = "prob7_u_cube_noV"
probsum3 = cx_mat()
probsum3.load(path)
```

```python
print(probsum1.min(),probsum1.max())
```

```python
plt.plot(abs(probsum1-1))
#plt.yscale('log')
```

```python
plt.plot(abs(probsum3-1))
#plt.yscale('log')
```

```python
import matplotlib.gridspec as gridspec

h = 2.5e-5
t = [h*i for i in range(len(probsum1))]

fig = plt.figure(figsize = (10,6))
gs1 = gridspec.GridSpec(1, 1, hspace=0)
gs1.update(wspace=0.0, hspace=0.0) # set the spacing between axes. 

#axs = plt.subplots(gs1[0],gs1[1], sharey=False)

axs = [plt.subplot(gs1[0])]
ax = axs[0]
linewidth = 2

ax.plot(t,abs(probsum1-1), color='tab:green', alpha=0.95, linestyle='dashed', label ='No barrier',linewidth =linewidth)
ax.plot(t, abs(probsum3-1), color='tab:blue', alpha=0.95, linestyle='solid', label ='Double slit barrier',linewidth =linewidth)

ax.set_facecolor((0.98, 0.98, 0.98))

ax.set_xlim(0,0.008)
#ax.set_ylim(1.85e-23,3.6e-23)
ax.set_xlabel(r'$t$')                                
ax.set_ylabel(r'$|\delta P |\:$', rotation=0, labelpad=18)

font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 20}
plt.rc('font', **font)



ax.legend(loc=(0.1,0.6))

plt.tight_layout()
plt.savefig('prob7.pdf')
fig.show()
```
